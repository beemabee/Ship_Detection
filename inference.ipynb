# `Inference on Video`

```python
# import Libraries

from ultralytics import YOLO
from PIL import Image
```

### `Video_1`

```python
# load small model (yolov8s) 
model_small = YOLO('./Ship Detection/First_Training_Smallpt/weights/best.pt')

# load path video
source = 'inference_1.mp4'
```

```python
# run inference
results = model_small(source, 
                      stream=True,
                      save=True)

# show results
for r in results:
    boxes = r.boxes
    masks = r.masks
    keypoints = r.keypoints
    probs = r.probs
```
